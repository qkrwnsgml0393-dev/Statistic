```python
import pandas as pd
df = pd.read_csv('KS_NFA_FTNESS_MESURE_ITEM_MESURE_INFO_202401.csv')
df.head()
```

```python
map_col = {
     'MESURE_IEM_001_VALUE' : 'hight',
     'MESURE_IEM_002_VALUE' : 'weight',
     'MESURE_IEM_003_VALUE' : 'fatrate',
     'MESURE_IEM_004_VALUE' : 'waist',
     'MESURE_IEM_005_VALUE' : 'diastolic_bp',
     'MESURE_IEM_006_VALUE' : 'systolic_bp',
     }

df = df.rename(columns=map_col)
display(df.head())
```

```python
sample_df = df.iloc[0:, 9:15]
```

```python
sample_df.describe()
```

```python
sample_df.var()
sample_df.std()
sample_df.mode()
sample_df.quantile(0.25)
```

```python
sample_df.isnull().sum()
```

```python
null_data = sample_df.isnull().any(axis=1)
```

```python
f_df = sample_df[~null_data]
f_df
```

```python
f_df.dtypes
```

```python
Q1 = f_df.quantile(0.25)
Q3 = f_df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
print(lower)
print(upper)
```

```python
summary = pd.DataFrame({
    'min' : f_df.min(),
    'Q1' : Q1,
    'Q3' : Q3,
    'lower' : lower,
    'uppper' : upper,
    'max' : f_df.max()
})
print(summary)
```

```python
#outlier = (f_df<lower) | (f_df>upper)를 쓰면 두 객체의 이덱스 기준이 다르다 -> False

#outlier = (f_df[f_df.columns] < lower[f_df.columns]) | (f_df[f_df.columns] > upper[f_df.columns])
# 이상치 마스크 (컬럼 기준 비교)
outlier_mask = f_df.lt(lower, axis=1) | f_df.gt(upper, axis=1)

# 이상치가 하나라도 있는 행만 필터링
outlier_rows = f_df[outlier_mask.any(axis=1)]

print(f"이상치 행 개수: {len(outlier_rows)}")
outlier_rows

```

```python
main_df = f_df[~outlier_mask.any(axis=1)]
main_df
```

```python
main_df.describe()
```

```python
import matplotlib.pyplot as plt
main_df[['hight',
      'weight',
      'fatrate',
      'waist',
      'diastolic_bp',
      'systolic_bp'
      ]].hist(bins=30, figsize=(10,6))
plt.show()

main_df[['hight',
      'weight',
      'fatrate',
      'waist',
      'diastolic_bp',
      'systolic_bp'
      ]].plot(kind='box', figsize=(8,5))
plt.show()
```

```python
before_stats = f_df.describe()
after_stats = main_df.describe()

before_stats.columns = ['before_' + col for col in before_stats.columns]
after_stats.columns = ['after_' + col for col in after_stats.columns]

compare_stats = pd.concat([before_stats, after_stats], axis=1)

diff_stats = after_stats.values - before_stats.values
diff_stats_df = pd.DataFrame(diff_stats,
                             columns=after_stats.columns,
                             index=after_stats.index)
final_stats = pd.concat([compare_stats, diff_stats_df], axis=1)
final_stats
```

```python
# 전/후 요약 (이미 있음)
before = f_df.describe()
after  = main_df.describe()

# 차이(후-전) 계산: 라벨 정합 자동
diff = after.sub(before)

# 1) 우리가 관심 있는 행만 추출 (평균/표준편차/최대)
focus = diff.loc[['mean','std','max']]

# 2) 절대값 기준으로 변화가 큰 항목 Top-10
big_changes = (focus.abs()
               .stack()                 # (지표, 변수) 멀티인덱스
               .sort_values(ascending=False)
               .head(10))

print("변화가 큰 Top-10 (지표, 변수)와 변화량:")
print(big_changes)
```

```python
!apt-get install -y fonts-nanum
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

# 1. 한글 폰트 설치
!apt-get update -qq
!apt-get install -y fonts-nanum

mpl.rc('font', family='NanumGothic')  # 나눔고딕 적용
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 부호 깨짐 방지

# 전/후 요약
before = f_df.describe()
after  = main_df.describe()

# 변화율 계산 (후-전) / 전
change_rate = after.sub(before).div(before) * 100  # %

# 관심 있는 지표만 추출
focus_metrics = ['mean', 'std', 'max']
change_focus = change_rate.loc[focus_metrics]

# 히트맵 시각화
plt.figure(figsize=(10, 5))
sns.heatmap(change_focus, annot=True, fmt=".1f", cmap="RdYlGn_r", center=0)
plt.title("이상치 제거 전후 변화율 (%)", fontsize=14)
plt.ylabel("통계 지표")
plt.xlabel("변수")
plt.show()
```

	•	결과적으로 이상치 제거는 분산 안정화와 신뢰도 개선에 기여했으며, 이후 가설검정/모델링에서 추정 오차 감소와 과대해석 방지 효과가 기대된다.

🔎 상세 해석 (포트폴리오용 본문)

이번 실험에서는 IQR(1.5×IQR) 기준으로 이상치를 제거하고, 전·후 변화율 히트맵(행: mean/std/max, 열: 변수)을 통해 영향도를 점검했다.
히트맵에서 **진한 음색(음수)**은 감소(안정화), **진한 양색(양수)**은 증가(확대)를 의미한다.
	1.	혈압 지표의 왜곡 제거

	•	systolic_bp의 **최댓값(max)**이 크게 하향되고, **표준편차(std)**가 급감했다. 이는 극단적 고값이 평균 주변의 분산을 과대추정하게 만들고 있었음을 나타낸다.
	•	평균 변화(mean)는 상대적으로 작아, 중심값은 유지되면서 흩어짐만 안정된 형태다. 실제 분석/예측 단계에서 이는 신뢰구간 축소와 검정력 향상으로 이어질 수 있다.

	2.	신체 치수의 상단 꼬리 정리

	•	weight, waist, fatrate, hight에서도 max가 유의하게 감소했다. 이는 측정/입력 오류 또는 비정상적으로 큰 값이 제거되었음을 시사한다.
	•	다만 해당 변수들의 std 감소폭은 혈압만큼 크지 않아, 일부 극단값 컷 수준으로 해석된다(중심경향 왜곡은 제한적).

	3.	실무적 영향

	•	이상치 제거는 평균 자체를 크게 바꾸기보다, 분산 안정화와 최댓값 클리핑을 통해 **모형 안정성(robustness)**과 지표 신뢰도를 높인다.
	•	특히 고혈압 분류·위험도 예측과 같이 임상 의사결정과 연결되는 문제에서, 극단치가 잔존하면 모델이 스케일에 과민해질 수 있으므로 사전 클리닝이 필수다.

⚠️ 주의 & 한계
	•	임상적 ‘이상값’ vs 통계적 ‘이상치’: 혈압, 허리둘레 등은 **도메인 기준(예: SBP≥140, DBP≥90)**과 통계 기준이 다를 수 있다. 두 기준을 함께 보고 최종 제거/보정 여부를 결정해야 한다.
	•	전체 vs 하위그룹: 전체 데이터에서 이상치가 없더라도, 성별/연령대/측정연도별로 보면 이상치가 나타날 수 있다. 그룹별 IQR/Robust Z 점검을 병행하는 것이 바람직하다.
	•	Winsorizing 대안: 삭제가 부담스럽다면 상·하위 p%를 윈저라이징(값을 경계로 눌러줌) 하는 보수적 대안도 고려 가능하다.
